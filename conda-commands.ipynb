# Note :- use "!" before conda & pip command in Jupyter Notebook

### List All Conda Packages 

```python
! conda list
```

### Upgrade one Conda Packages
-  conda upgrade package_name

### Upgrade all Conda packages
-  conda upgrade conda
-  conda upgrade --all

### Install a conda Package
-  conda install package_name like conda install numpy

### Install Multiple Packages
-  conda install numpy pandas

### Install Packages with Specific Version
-  conda numpy version=1.10

### Remove Package Name
-  conda remove package_name

### Search Package Name 
-  conda search *search_term*
-  conda search *beautifulsoup*
